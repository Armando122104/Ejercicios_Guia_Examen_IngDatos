Ejercicio 2: Transformación de Datos - Columnas Calculadas
Objetivo: Crear nuevas columnas basadas en cálculos
21.	Usando ventas_mensuales.xlsx, crea una columna llamada 'Total_Venta' que multiplique Cantidad por Precio_Unitario
22.	Calcula el total de ventas (suma de la columna Total_Venta)
23.	Encuentra el mes con mayor venta total
24.	Guarda el DataFrame modificado como 'ventas_procesadas.xlsx'


```python
# Leer el archivo
df = pd.read_excel('ventas_mensuales.xlsx')

# Columna de venta total
df['Total_Venta'] = df['Cantidad'] * df['Precio_Unitario']

# Calcular la suma del total de ventas
total_ventas = df['Total_Venta'].sum()
print(f'Total de ventas: ${total_ventas:.2f}')

# Mes con el max de ventas
mes_mayor = df.loc[df['Total_Venta'].idxmax(), 'Mes']
print(f'Mes con mayor venta: {mes_mayor}')

# Guardar el dataframe
df.to_excel('ventas_procesadas.xlsx', index=False)
```

Ejercicio 4: Agrupación y Agregación
Objetivo: Resumir datos por grupos
30.	Usando ventas_mensuales.xlsx, agrupa por 'Region' y calcula el total de ventas por región
31.	Agrupa por 'Producto' y encuentra la cantidad promedio vendida
32.	Usando empleados.xlsx, calcula el salario promedio por departamento
33.	Encuentra el salario máximo y mínimo por ciudad



```python
df = pd.read_excel('ventas_mensuales.xlsx')

# Calcular ventas totales (Cantidad * Precio_Unitario)
df['Venta_Total'] = df['Cantidad'] * df['Precio_Unitario']

# Agrupar por región y sumar ventas
ventas_por_region = df.groupby('Region')['Venta_Total'].sum().reset_index()
print("Total de ventas por región:")
print(ventas_por_region)
print('')

df2 = pd.read_excel('empleados.xlsx')

# Calcular el salario promedio por departamento
salario_promedio = df2.groupby('Departamento')['Salario'].mean().reset_index()
print("Salario promedio por departamento:")
print(salario_promedio)
print('')


# Encuentra el salario maximo y minimo por ciudad
salario_max_min = df2.groupby('Ciudad')['Salario'].agg(['max', 'min']).reset_index()
print("Salario máximo y mínimo por ciudad:")
print(salario_max_min)
```

Ejercicio 6: Unión de Datasets (Merge)
Objetivo: Combinar múltiples fuentes de datos
40.	Carga ventas_mensuales.xlsx y productos.xlsx
41.	Nota: Necesitarás crear una columna de unión. En ventas, mapea 'Producto' a un SKU simplificado
42.	Combina ambos DataFrames para obtener información completa de productos
43.	Calcula el stock restante después de las ventas


```python
ventas = pd.read_excel('ventas_mensuales.xlsx')
productos = pd.read_excel('productos.xlsx')

# Columna de unión (mapeo simplificado)
mapeo_producto_sku = {
    'Laptop': 'LAP001',
    'Mouse': 'MOU001',
    'Teclado': 'TEC001',
    'Monitor': 'MON001'
}

ventas['SKU'] = ventas['Producto'].map(mapeo_producto_sku)
print("Ventas con SKU: ")
print(ventas[['Mes', 'Producto', 'SKU', 'Cantidad']])
print('')

# Combinar DataFrames
ventas_completo = pd.merge(ventas, productos, left_on='SKU', right_on='SKU', how='left')
print("Ventas con informacion de productos: ")
print(ventas_completo[['Mes', 'Producto', 'Cantidad', 'Nombre_Producto', 'Stock']].head(10))
print('')

# Calcular stock restante
ventas_totales = ventas_completo.groupby(['SKU', 'Nombre_Producto', 'Stock']).agg({
    'Cantidad': 'sum'
}).reset_index()

ventas_totales['Stock_Restante'] = ventas_totales['Stock'] - ventas_totales['Cantidad']
ventas_totales['Stock_Restante'] = ventas_totales['Stock_Restante'].clip(lower=0)

print("Stock restante despues de ventas: ")
print(ventas_totales[['SKU', 'Nombre_Producto', 'Stock', 'Cantidad', 'Stock_Restante']])

```

Ejercicio 7: Pipeline ETL Completo
Objetivo: Crear un proceso ETL de principio a fin
Crea un script que:
44.	Extract: Cargue ventas_mensuales.xlsx
45.	Transform:
   - Calcule Total_Venta (Cantidad * Precio_Unitario)
   - Filtre solo ventas mayores a 1000
   - Agrupe por Región y calcule suma de ventas
   - Ordene de mayor a menor
46.	Load: Guarde el resultado en 'reporte_regional.xlsx'


```python
ventas = pd.read_excel('ventas_mensuales.xlsx')

# Calcular total_ventas
ventas['Total_Venta'] = ventas['Cantidad'] * ventas['Precio_Unitario']
print('Total de ventas: ', ventas['Total_Venta'])
print('')

# Filtrar solo ventas mayores a 1000
ventas = ventas[ventas['Total_Venta'] > 100]
print('Filtrado de ventas mayores a 1000: ',ventas)
print('')

# Agrupar por region y calcular la suma
ventas_por_region = ventas.groupby('Region')['Total_Venta'].sum().reset_index()
print('Agrupado por region y suma total: ', ventas_por_region)
print('')

# Ordenar de mayor a menor
ventas_por_region = ventas_por_region.sort_values(by='Total_Venta', ascending=False)
print('Ordenado de mayor a menor: ', ventas_por_region)

# Guarde el resultado en 'report_regional.xlsx'
reporte_regional = ventas_por_region.to_excel('reporte_regional.xlsx', index=False)
```
