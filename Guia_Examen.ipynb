```python
#Usando ventas_mensuales.xlsx, escribe el código para crear una columna 'Ingreso_Total' que sea el producto de Cantidad por Precio_Unitario.

df = pd.read_excel('ventas_mensuales.xlsx')

df['Ingreso_Total'] = df['Cantidad'] * df['Precio_Unitario']

print(df)
```

```python
#Escribe el código para filtrar solo los empleados del departamento 'Ventas' con salario mayor a 35000.

df = pd.read_excel('empleados.xlsx')

datos_filtrados = df[(df['Departamento'] == 'Ventas') & (df['Salario'] > 35000)]
print(datos_filtrados)
```

```python
 #Usando ventas_mensuales.xlsx, agrupa por 'Region' y calcula la suma total de ventas (Cantidad * Precio_Unitario) por región.

df = pd.read_excel('ventas_mensuales.xlsx')

df['Ingreso_Total'] = df['Cantidad'] * df['Precio_Unitario']

datosAgrupados = df.groupby('Region')['Ingreso_Total'].sum()
print(datosAgrupados)
```

```python
#Usando datos_con_problemas.xlsx, #escribe el código para: (a) identificar valores nulos en la columna 'Monto',
#(b) eliminar esas filas, y (c) guardar el resultado en 'datos_procesados.xlsx'.

df = pd.read_excel('datos_con_problemas.xlsx')

df = df.dropna(subset=['Monto'])
print(df)

df.to_excel('datos_procesados.xlsx', index=False)
```
